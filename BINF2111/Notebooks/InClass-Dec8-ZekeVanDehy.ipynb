# The sin() function

To plot a sin wave we can import the sin() function which is part of the numpy package. Below is an example.

```python
import numpy as np

X = range(10)
Y = []

for i in X:
    Y.append(np.sin(X[i]))

plt.plot(X,Y)
```

# Can it look better?

In the cell above, there is a plot of a sin wave but it looks kind of choppy. This is caused by the fact that there are only a few points for each cycle of the wave. Recall that a sin wave repeates every $2\pi$ radians. Since $\pi = 3.14$ radians and $2\pi = 6.28$ radians, there are only a little over 6 points in every cycle of the wave in the plot below. 6 points is not enough to really define the shape of a sin wave.

Can you find a way to modify the plot below to make the wave look more smooth? Hint: there needs to be more points on the plot per cycle.

```python
X = [x/100 for x in range(1000)]
Y = []

for i in X:
    Y.append(np.sin(i))

plt.plot(X,Y)
```

# Using a comprehension inside a plt.plot() call

We can put a comprehension inside a call tp plot.plot() instead of having a list of values outside the function.

```python
X = range(1000)

plt.plot(X,[np.sin(i/100) for i in X])
```

# the linspace() function

When you are plotting some data, you often have a very specific range of values which you want to plot over on the x axis. The linspace() function gives you mor control over its values than the range() function.

linspace uses the following syntax:

linspace(start_value, end_value, number_of_steps)
    
For example, if I wanted to make the values of x go from 0 to 10 but take 1000 steps in the process I could use:

X = linspace(0,10,1000)

Knowing that a sin wave repeats about every 6.29 radians, make a smooth plot of a sin wave which shows only a single cycle of the wave. This means that the wave will start at some y value and stop as soon as it returns to that value.

```python
X = np.linspace(0,6.29,1000)
Y = np.sin(X)

plt.plot(X,Y)
```

# Plotting multiple things on the same graph

More than one thing can be plotted at the same time. Below is an exampe of two waves which are 180 degrees out of sync with eachother.

By adding a fixed value to the argument of a sin() function, it can be moved left or right by that value. Since 180 degrees = $\pi$ radians, the waves can be made to have a different phase of 180 degrees by adding 3.14 to the argument inside one of the sin() functions.

```python
X =range(1000)
Y1 = [np.sin(i/100) for i in X]
Y2 = [np.sin(i/100+3.14) for i in X]

plt.plot(X,Y1)
plt.show()
plt.plot(X,Y2)
plt.show()
plt.plot(X,Y1)
plt.plot(X,Y2)
```

# Plotting a wave interference pattern

In nature, waves often interfere with eachother. When this happens, the amplitudes of the waves add together at each point. Depending on the phase of the interfering waves, they can either interfere constructively or destructively. For constructive interference, the amplitude of the resulting wave is greater than that of the original waves. For destructive interference, the amplitude of the resulting combined wave has a reduced amplitude.

In the cell below, use two sin waves with amplitudes of one to generate a series of interference waves with amplitudes close to the following values: 0, 0.5, 1, 1.5 and 1.75. 

```python
import math
X =range(1000)
Y1 = [np.sin(i/100) for i in X]
Y2 = [np.sin(i/100+math.pi) for i in X]

cycle = 2*math.pi
for y in range(100):
    Y2 = [np.sin(i/100+(cycle*y/10)) for i in X]
    YC = [Y1[i]+Y2[i] for i in X]
    plt.plot(X, YC)

# plt.ylim(-1.0, 1.0)

```
