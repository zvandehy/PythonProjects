```python
from Bio import AlignIO
```

```python
align = AlignIO.read("PF09395_seed.sth", "stockholm") #filename, format
print("# Sequences: {}".format(len(align)))
print("Alignment Length: {}".format(align.get_alignment_length()))

for record in align:
    print(record.seq + " " + record.id)
```

```python
#how often do different substitutions occur?
print(align.substitutions)
#amino acid stays in A, 197 times out of all of these pairwise comparisons
#T changed to M 9 times
print(align.substitutions/len(align))
```

```python
#convert file format
print(align.format("fasta"))
AlignIO.convert("PF09395_seed.sth", "stockholm", "PF09395_seed.phy", "phylip")
```

```python
#BLAST sequence search tool finds regions of similarity between biological sequences
#search all of the databases from NCBI
#rna, dna, protein sequences
#returns a file that tells you how the search results matched up with your query
#what database records hit, the name of that record, and the sequence and how your sequence aligns with it
#my_blast.xml is an example search result file

#Bio.SearchIO provides methods for reading these search results
#list of high scoring segment pairs

from Bio import SearchIO
result = SearchIO.read("my_blast.xml", "blast-xml")
print(result)
```

```python
for hit in result.hits[:5]:
    print(hit)
```

```python
for hit in result.hits[:5]:
    for hsp in hit.hsps:
        print(hsp.aln.format("fasta"))
```

```python
blatresult = SearchIO.read("my_blat.psl", "blat-psl")
for hit in blatresult.hits[:5]:
    print(hit)
    for hsp in hit.hsps:
        print(hsp)
```

```python
SearchIO.convert("my_blast.xml", "blast-xml", "my_blast.txt", "blast-tab")
```
